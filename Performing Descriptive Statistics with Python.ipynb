# 7. Calculate the z score of the following numbers:

X = [4.5,6.2,7.3,9.1,10.4,1
1]

```python
import numpy as np
X = [4.5,6.2,7.3,9.1,10.4,11]

#To calulate Z score : (X - mean)/SD

#Mean
X_mean = np.mean(X)

#Std_dev
X_SD = np.std(X)

for X in X:
    Z = (X - X_mean)/X_SD,
    print("Z score of ", {X},"is", Z)
```

# 20. Using the z-score of the income variable, find out the number of observations outside the +-3σ.

```python
from scipy.stats import zscore

# Calculate the z-scores for the income variable
data['Income_zscore'] = zscore(data['Income'])

# Count the number of observations outside the ±3σ range
outliers = data[(data['Income_zscore'] < -3) | (data['Income_zscore'] > 3)]
nnum_outliers = len(outliers)

print('Number of observations outside the ±3σ range:', num_outliers)
```
